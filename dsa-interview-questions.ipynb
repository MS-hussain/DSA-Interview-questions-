Q1:How do you find all duplicate number on a given integer array also what are there frequency

```python
def find_duplicates_and_frequency(arr):
    frequency ={}
    repeted_num = {}
    for i in arr:
        if i in frequency:
            frequency[i] += 1
        else:
            frequency[i] = 1
    print(frequency)
    for num,count in frequency.items():
        if count>1:
            repeted_num[num]=count
        return repeted_num
    
# Example Usage:
numbers = [1, 2, 3, 4, 2, 5, 6, 1, 7, 7, 2]
result = find_duplicates_and_frequency(numbers)
print(f"Duplicate numbers and their frequencies: {result}")
```

Q2:How do you find the missing number in a given integer array of 1 to N

```python
def missing_value(arr,N):
    actual_sum=sum(arr)
    expected_sum=(N*(N+1)//2)
    missing =expected_sum-actual_sum
    return missing 
arr =[1,2,3,4,5,6,8,9]
N = 9
result=missing_value(arr,N)
print("the missing value in the set is {}".format(result))
```

Q3:How do you find the largest and smallest number in an unsorted integer array?

```python
def MAX_and_MIN(arr):
    max = arr[0]
    min= arr[0]
    for num in arr:
        if num > max:
            max = num
        if num < min:
            min = num
    return min,max
list = [1,2,3,11,22,33,44,55,66,0,100,-7,240]
min,max = MAX_and_MIN(list)
print("the maximum and minimum number in the list are {} and {} respectively".format(max,min))
```

Q4:How do you find all pairs of an integer array whose sum is equal to a given number?

```python
def find_pairs_with_sum(nums, target_sum):
    seen = set()
    result = []

    for num in nums:
        complement = target_sum - num
        if complement in seen:
            result.append((num, complement))
        seen.add(num)
    
    return result

# Example Usage:
numbers = [1, 2, 3, 4, 5, 6,10,11,27]
target_sum = 9
result = find_pairs_with_sum(numbers, target_sum)
print(f"Pairs with sum {target_sum}: {result}")
```

Q5:How are duplicates removed from a given array

```python
def remove_duplicates(nums):
    unique_set = set()
    result = []

    for num in nums:
        if num not in unique_set:
            unique_set.add(num)
            result.append(num)

    return result

# Example Usage:
numbers = [1, 2, 3, 2, 4, 5, 6, 1, 7, 8, 9, 5]
result = remove_duplicates(numbers)
print(f"Array after removing duplicates: {result}")
```

Q6:How is an integer array sorted in place using the quicksort algorithm?

```python
def quicksort(nums, low, high):
    if low < high:
        # Find the partition index
        partition_index = partition(nums, low, high)

        # Recursively sort the sub-arrays
        quicksort(nums, low, partition_index)
        quicksort(nums, partition_index + 1, high)

def partition(nums, low, high):
    pivot = nums[(low + high) // 2]
    i = low - 1
    j = high + 1

    while True:
        i += 1
        while nums[i] < pivot:
            i += 1

        j -= 1
        while nums[j] > pivot:
            j -= 1

        if i >= j:
            return j

        # Swap nums[i] and nums[j]
        nums[i], nums[j] = nums[j], nums[i]

# Example Usage:
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
quicksort(numbers, 0, len(numbers) - 1)
print("Sorted array using quicksort:", numbers)
```

How do you print duplicate characters from a string

```python
def print_duplicates(s):
    char_count = {}

    for char in s:
        char_count[char] = char_count.get(char, 0) + 1

    print("Duplicate characters in '{}' are:".format(s))
    for char, count in char_count.items():
        if count > 1:
            print("{} ({} times)".format(char, count))

# Example Usage:
input_string = "hello"
print_duplicates(input_string)

```

How do you check if two strings are anagrams of each other?

```python
def are_anagrams(str1, str2):
    # Remove spaces and convert to lowercase for case-insensitive comparison
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    if len(str1) != len(str2):
        return False

    char_count = {}

    for char in str1:
        char_count[char] = char_count.get(char, 0) + 1

    for char in str2:
        if char not in char_count or char_count[char] == 0:
            return False
        char_count[char] -= 1

    return True

# Example Usage:
string1 = "Listen"
string2 = "Silent"
result = are_anagrams(string1, string2)
print(f"Are '{string1}' and '{string2}' anagrams? {result}")
```

How do you count the number of vowels and consonants in a given string?

```python
def count_vowels_and_consonants(s):
    vowels = "aeiouAEIOU"
    vowel_count = consonant_count = 0

    for char in s:
        if char.isalpha():
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    return vowel_count, consonant_count

# Example Usage:
input_string = "Hello, World!"
vowels, consonants = count_vowels_and_consonants(input_string)
print(f"Number of vowels: {vowels}, Number of consonants: {consonants}")
```

How do you check if two strings are a rotation of each other? 

```python
def are_rotations(str1, str2):
    if len(str1) != len(str2):
        return False

    concatenated_str = str1 + str1

    return str2 in concatenated_str

# Example Usage:
string1 = "abcde"
string2 = "cdeab"
result = are_rotations(string1, string2)
print(f"Are '{string1}' and '{string2}' rotations of each other? {result}")
```

How do you check if a given string is a palindrome?

```python
def is_palindrome(s):
    s = s.lower()  # Convert to lowercase for case-insensitive comparison
    start, end = 0, len(s) - 1

    while start < end:
        while start < end and not s[start].isalnum():
            start += 1
        while start < end and not s[end].isalnum():
            end -= 1

        if s[start] != s[end]:
            return False

        start += 1
        end -= 1

    return True

# Example Usage:
input_string = "A man, a plan, a canal, Panama"
result = is_palindrome(input_string)
print(f"Is '{input_string}' a palindrome? {result}")
```
